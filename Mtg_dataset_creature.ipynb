```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt 
import matplotlib.ticker as mtick

pd.plotting.register_matplotlib_converters()
%matplotlib inline
import seaborn as sns
```

```python
pagina1 = requests.get("https://gatherer.wizards.com/Pages/Search/Default.aspx?type=+[Creature]")

mtg = BeautifulSoup(pagina1.content, "html.parser")

nome = mtg.find_all(class_="cardTitle")
mana = mtg.find_all(class_="convertedManaCost")
tipo = mtg.find_all(class_="typeLine")
conteudo = mtg.find_all(class_="rulesText")

valor = ""
tuti = ""
geti = ""
yeti = ""


for ml in range(99):
  geti = ""
    
  valor += nome[ml].get_text()

  valor += " "

  valor += mana[ml].get_text()

  valor += " "

  tuti = tipo[ml].get_text()
  tuti = tuti.replace("                                ", " ")
  tuti = tuti.replace("\t", "")
  tuti = tuti.replace("\r", "")
  tuti = tuti.replace("\n", "")
  valor += tuti

  valor += " "

  geti = geti.join([img['alt'] for img in conteudo[ml].find_all('img', alt=True)])
  geti = geti.replace("\t", "")
  geti = geti.replace("\r", "")
  geti = geti.replace("\n", "")
  valor += geti

  valor += " "

  yeti = conteudo[ml].get_text()
  yeti = yeti.replace("\t", "")
  yeti = yeti.replace("\r", "")
  yeti = yeti.replace("\n", "")
  valor += yeti
  


for numero_paginas in range (115):
  # O número colocado aqui  ^^^^^ define quantas páginas + 1 vão ser usadas para fazer o dataset (cada página tem 100 cartas)

  x = "https://gatherer.wizards.com/Pages/Search/Default.aspx?page=" + str(numero_paginas + 1) + "&type=+[Creature]"

  pagina2 = requests.get(x)

  mtg2 = BeautifulSoup(pagina2.content, "html.parser")

  nome = mtg2.find_all(class_="cardTitle")
  mana = mtg2.find_all(class_="convertedManaCost")
  tipo = mtg2.find_all(class_="typeLine")
  conteudo = mtg2.find_all(class_="rulesText")

  for ml in range(99):
    geti = ""
      
    valor += nome[ml].get_text()

    valor += " "

    valor += mana[ml].get_text()

    valor += " "

    tuti = tipo[ml].get_text()
    tuti = tuti.replace("                                ", " ")
    tuti = tuti.replace("\t", "")
    tuti = tuti.replace("\r", "")
    tuti = tuti.replace("\n", "")
    valor += tuti

    valor += " "

    geti = geti.join([img['alt'] for img in conteudo[ml].find_all('img', alt=True)])
    geti = geti.replace("\t", "")
    geti = geti.replace("\r", "")
    geti = geti.replace("\n", "")
    valor += geti

    valor += " "

    yeti = conteudo[ml].get_text()
    yeti = yeti.replace("\t", "")
    yeti = yeti.replace("\r", "")
    yeti = yeti.replace("\n", "")
    valor += yeti
```

```python
print(valor)
```

```python
with open("MTG-DATASET.txt", "a") as file_object:
    file_object.write(valor)
```
